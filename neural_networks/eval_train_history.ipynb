```python
import pandas as pd
import ast
```

```python
BASEDIR = 'out/'
HISTORY_LINE_NUMBER = 32
```

```python
# extracts a training history from the given out file 
def extract_history(str: file_name):
    history = None
    with open(file_name, 'r') as f:
        lines=f.readlines()
        history = lines[HISTORY_LINE_NUMBER]
        
    history_trimmed = history[9:]  # remove prefix "History: "
    history_dict = ast.literal_eval(history_trimmed)
    return history_dict
```

```python
# plots a chart with losses during the training epochs
def plot_loss(str: out_file):
    history = extract_history(out_file)
    loss_df = pd.DataFrame(history)
    loss_df.plot(figsize=(12,8))
```
